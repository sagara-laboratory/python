```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,200)
```

```python
y = np.linspace(-5,5,200)
```

```python
X,Y = np.meshgrid(x,y)
```

```python
Z=X.ravel()**2-Y.ravel()**2
```

```python
plt.contour(X,Y,Z.reshape(X.shape),colors="k")
```

```python
plt.show()
```
