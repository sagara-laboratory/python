```python
import pandas as pd
```

```python
import numpy as np
```

```python
wine = pd.read_csv("winequality-red.csv", sep=";")
```

```python
from sklearn import linear_model
```

```python
clf = linear_model.LinearRegression()
```

```python
X = wine.loc[:, ['density']].as_matrix()
```

```python
Y = wine['alcohol'].as_matrix()
```

```python
clf.fit(X, Y)
```

```python
print(clf.coef_)
```

```python
print(clf.intercept_)
```

```python
print(clf.score(X, Y))
```
