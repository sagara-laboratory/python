```python
import random

def bubble_sort(data, n):
    for i in range(n):                                      
        for j in range(n-1, i, -1):                         
            if data[j-1] > data[j]:                         
                data[j], data[j-1] = data[j-1], data[j]
            print('i=', i, 'j=', j, data)
    return data
```

```python
y = random.sample(range(5), k=5)
print('initial array=', y)
y = bubble_sort(y, len(y))
```
